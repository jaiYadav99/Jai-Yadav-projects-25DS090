```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```


# IMPORT LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt




# NUMPY (Numerical Operations)
print("----- NUMPY -----")

# Create array
arr = np.array([10, 20, 30, 40, 50])
print("Array:", arr)

# Basic operations
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))

# Reshape array
arr2 = np.arange(1, 10).reshape(3, 3)
print("2D Array:\n", arr2)


# PANDAS (Data Handling)

print("\n----- PANDAS -----")

# Create DataFrame
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, 90, 78, 92, 88],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

# Display data
print("DataFrame:\n", df)

# Basic operations
print("\nHead:\n", df.head())
print("\nInfo:")
df.info()

print("\nDescribe:\n", df.describe())

# Filtering
print("\nStudents with Marks > 85:\n", df[df["Marks"] > 85])

# Add new column
df["Grade"] = ["B", "A", "C", "A", "B"]
print("\nUpdated DataFrame:\n", df)

# MATPLOTLIB (Visualization)

print("\n----- MATPLOTLIB -----")

# Line Plot
plt.figure()
plt.plot(df["Name"], df["Marks"], marker="o")
plt.title("Marks Line Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.tight_layout()
plt.show()

# Bar Chart
plt.figure()
plt.bar(df["Name"], df["Marks"], color="skyblue")
plt.title("Marks Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.tight_layout()
plt.show()

# Scatter Plot
plt.figure()
plt.scatter(df["Age"], df["Marks"], color="red")
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.tight_layout()
plt.show()

# Histogram
plt.figure()
plt.hist(df["Marks"], bins=5, color="green", edgecolor="black")
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```

# IMPORT LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn style
sns.set()

# NUMPY (Numerical Operations)

print("----- NUMPY -----")

arr = np.array([10, 20, 30, 40, 50])
print("Array:", arr)

print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))

arr2 = np.arange(1, 10).reshape(3, 3)
print("2D Array:\n", arr2)

# PANDAS (Data Handling)
print("\n----- PANDAS -----")

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, 90, 78, 92, 88],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

print("DataFrame:\n", df)
print("\nDescribe:\n", df.describe())

# SEABORN VISUALIZATION

print("\n----- SEABORN -----")

# Line Plot
plt.figure()
sns.lineplot(x="Name", y="Marks", data=df)
plt.title("Marks Line Chart")
plt.tight_layout()
plt.show()

# Bar Plot
plt.figure()
sns.barplot(x="Name", y="Marks", data=df)
plt.title("Marks Bar Chart")
plt.tight_layout()
plt.show()

# Scatter Plot
plt.figure()
sns.scatterplot(x="Age", y="Marks", data=df)
plt.title("Age vs Marks")
plt.tight_layout()
plt.show()

# Histogram
plt.figure()
sns.histplot(df["Marks"], kde=True)
plt.title("Marks Distribution")
plt.tight_layout()
plt.show()

# Box Plot
plt.figure()
sns.boxplot(x="Marks", data=df)
plt.title("Box Plot of Marks")
plt.tight_layout()
plt.show()

# HEATMAP (Correlation)

plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
